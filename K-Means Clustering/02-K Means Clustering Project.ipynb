## K Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=2, max_iter=500)
```

```python
X= college.drop('Private', axis=1)
kmean.fit(X)
```

```python
kmean.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters.

```python
college.Private.unique()
```

```python
college['Cluster'] = college['Private'].apply(lambda x: 1 if x=='Yes'else 0)
```

```python
print(college.Private.value_counts())
y = np.bincount(kmean.labels_)
ii = np.nonzero(y)[0]
a = list(zip(ii, y[ii]))
print(y)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(accuracy_score(college['Cluster'], kmean.labels_))
print(confusion_matrix(college['Cluster'], kmean.labels_))
print('\n')
print(classification_report(college['Cluster'], kmean.labels_))
```
